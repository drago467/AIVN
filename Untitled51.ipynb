```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
agri_df = pd.read_csv('Agri.csv')
bidv_df = pd.read_csv('BIDV.csv')
vcb_df = pd.read_csv('VCB.csv')
vietin_df = pd.read_csv('Vietin.csv')
```

```python
print("Agri DataFrame:")
print(agri_df.head())

print("\nBIDV DataFrame:")
print(bidv_df.head())

print("\nVCB DataFrame:")
print(vcb_df.head())

print("\nVietin DataFrame:")
print(vietin_df.head())
```

```python
# Concatenate all DataFrames into one
combined_df = pd.concat([agri_df[['money', 'bank']], bidv_df[['money', 'bank']],
                         vcb_df[['money', 'bank']], vietin_df[['money', 'bank']]])
```

```python
combined_df['money'] = combined_df['money'] / 1000
```

## **BOX PLOT**


```python
# Plot boxplot
plt.figure(figsize=(10, 6))
combined_df.boxplot(column='money', by='bank', grid=False)
plt.title('Boxplot of Money Transfers by Bank')
plt.suptitle('')  # Remove the default title to clean up the plot
plt.xlabel('Bank')
plt.ylabel('Money')
plt.show()
```

```python
# Loại bỏ các giá trị ngoại lệ dựa trên phần trăm thứ 5 và 95
lower_bound = combined_df['money'].quantile(0.05)
upper_bound = combined_df['money'].quantile(0.95)

# Lọc dữ liệu để chỉ lấy các giá trị trong khoảng 5% đến 95%
filtered_df = combined_df[(combined_df['money'] >= lower_bound) & (combined_df['money'] <= upper_bound)]

# Vẽ boxplot sau khi loại bỏ các giá trị ngoại lệ
plt.figure(figsize=(10, 6))
filtered_df.boxplot(column='money', by='bank', grid=False)
plt.title('Boxplot of Money Transfers by Bank (Without Outliers)')
plt.suptitle('')  # Bỏ tiêu đề phụ để làm gọn đồ thị
plt.xlabel('Bank')
plt.ylabel('Money')
plt.show()
```

```python
# Hàm vẽ boxplot cho từng ngân hàng
def plot_boxplot_for_bank(bank_df, bank_name):
    plt.figure(figsize=(6, 4))
    plt.boxplot(bank_df['money'], vert=True)
    plt.title(f'Boxplot of Money Transfers for {bank_name}')
    plt.ylabel('Money')
    plt.show()

# Tạo từng DataFrame riêng cho mỗi ngân hàng
agri_df = combined_df[combined_df['bank'] == 'Agri']
bidv_df = combined_df[combined_df['bank'] == 'BIDV']
vcb_df = combined_df[combined_df['bank'] == 'VCB']
vietin_df = combined_df[combined_df['bank'] == 'Vietin']

# Vẽ boxplot cho từng ngân hàng
plot_boxplot_for_bank(agri_df, 'Agri')
plot_boxplot_for_bank(bidv_df, 'BIDV')
plot_boxplot_for_bank(vcb_df, 'VCB')
plot_boxplot_for_bank(vietin_df, 'Vietin')
```

```python
# Hàm loại bỏ ngoại lệ dựa trên phần trăm thứ 5 và 95
def remove_outliers(bank_df):
    lower_bound = bank_df['money'].quantile(0.05)
    upper_bound = bank_df['money'].quantile(0.95)
    return bank_df[(bank_df['money'] >= lower_bound) & (bank_df['money'] <= upper_bound)]

# Hàm vẽ boxplot cho từng ngân hàng
def plot_boxplot_for_bank(bank_df, bank_name):
    plt.figure(figsize=(6, 4))
    plt.boxplot(bank_df['money'], vert=True)
    plt.title(f'Boxplot of Money Transfers for {bank_name} (Without Outliers)')
    plt.ylabel('Money')
    plt.show()

# Tạo từng DataFrame riêng cho mỗi ngân hàng và xử lý ngoại lệ
agri_df = combined_df[combined_df['bank'] == 'Agri']
agri_df_no_outliers = remove_outliers(agri_df)

bidv_df = combined_df[combined_df['bank'] == 'BIDV']
bidv_df_no_outliers = remove_outliers(bidv_df)

vcb_df = combined_df[combined_df['bank'] == 'VCB']
vcb_df_no_outliers = remove_outliers(vcb_df)

vietin_df = combined_df[combined_df['bank'] == 'Vietin']
vietin_df_no_outliers = remove_outliers(vietin_df)

# Vẽ boxplot cho từng ngân hàng sau khi xử lý ngoại lệ
plot_boxplot_for_bank(agri_df_no_outliers, 'Agri')
plot_boxplot_for_bank(bidv_df_no_outliers, 'BIDV')
plot_boxplot_for_bank(vcb_df_no_outliers, 'VCB')
plot_boxplot_for_bank(vietin_df_no_outliers, 'Vietin')
```

```python
# Tạo biểu đồ histogram cho từng ngân hàng
plt.figure(figsize=(10, 6))

# Vẽ biểu đồ histogram
for bank in combined_df['bank'].unique():
    subset = combined_df[combined_df['bank'] == bank]
    plt.hist(subset['money'], bins=20, alpha=0.5, label=bank)

plt.title('Histogram of Money Transfers by Bank')
plt.xlabel('Money')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
def plot_histogram(bank_df, bank_name):
    plt.figure(figsize=(8, 5))
    plt.hist(bank_df['money'], bins=20, alpha=0.7, color='blue')
    plt.title(f'Histogram of Money Transfers for {bank_name}')
    plt.xlabel('Money')
    plt.ylabel('Frequency')
    plt.show()

# Vẽ histogram cho từng ngân hàng
plot_histogram(agri_df, 'Agri')
plot_histogram(bidv_df, 'BIDV')
plot_histogram(vcb_df, 'VCB')
plot_histogram(vietin_df, 'Vietin')
```

```python
# Hàm vẽ biểu đồ histogram cho từng ngân hàng với tùy chỉnh số lượng bins
def plot_histogram(bank_df, bank_name, bins):
    plt.figure(figsize=(8, 5))
    plt.hist(bank_df['money'], bins=bins, alpha=0.7, color='blue')
    plt.title(f'Histogram of Money Transfers for {bank_name}')
    plt.xlabel('Money')
    plt.ylabel('Frequency')
    plt.show()

# Vẽ histogram cho từng ngân hàng với số lượng bins khác nhau
plot_histogram(agri_df, 'Agri', bins=15)   # Sử dụng 15 bins cho Agri
plot_histogram(bidv_df, 'BIDV', bins=20)   # Sử dụng 20 bins cho BIDV
plot_histogram(vcb_df, 'VCB', bins=25)     # Sử dụng 25 bins cho VCB
plot_histogram(vietin_df, 'Vietin', bins=30)  # Sử dụng 30 bins cho Vietin
```

```python
# Kiểm tra thống kê mô tả cho cột "money"
print(agri_df['money'].describe())
```

```python
# Loại bỏ giá trị lớn bất thường ở 99th percentile
upper_limit = agri_df['money'].quantile(0.99)
filtered_agri_df = agri_df[agri_df['money'] <= upper_limit]

# Vẽ lại histogram sau khi loại bỏ giá trị ngoại lệ
plot_histogram(filtered_agri_df, 'Agri', bins=15)

# Vẽ histogram với log scale trên trục x
plt.figure(figsize=(8, 5))
plt.hist(agri_df['money'], bins=15, alpha=0.7, color='blue')
plt.xscale('log')
plt.title('Histogram of Money Transfers for Agri (Log Scale)')
plt.xlabel('Money (Log Scale)')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu bạn cung cấp từ 4 ngân hàng (giả sử đã đọc từ CSV)
agri_df = pd.read_csv('Agri.csv')
bidv_df = pd.read_csv('BIDV.csv')
vcb_df = pd.read_csv('VCB.csv')
vietin_df = pd.read_csv('Vietin.csv')

# Đảm bảo cột "date" được chuyển sang định dạng ngày
agri_df['date'] = pd.to_datetime(agri_df['date'], format='%d/%m', errors='coerce')
bidv_df['date'] = pd.to_datetime(bidv_df['date'], format='%d/%m %H:%M:%S', errors='coerce')
vcb_df['date'] = pd.to_datetime(vcb_df['date'], format='%d/%m', errors='coerce')
vietin_df['date'] = pd.to_datetime(vietin_df['date'], format='%d/%m %H:%M:%S', errors='coerce')

# Hàm vẽ biểu đồ lượng tiền được chuyển qua từng ngày
def plot_money_by_day(bank_df, bank_name):
    # Nhóm dữ liệu theo ngày và tính tổng lượng tiền chuyển trong mỗi ngày
    daily_sum = bank_df.groupby(bank_df['date'].dt.date)['money'].sum()

    # Vẽ biểu đồ
    plt.figure(figsize=(10, 5))
    daily_sum.plot(kind='line', marker='o')
    plt.title(f'Total Money Transferred by Day for {bank_name}')
    plt.xlabel('Date')
    plt.ylabel('Total Money')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Vẽ biểu đồ cho từng ngân hàng
plot_money_by_day(agri_df, 'Agri')
plot_money_by_day(bidv_df, 'BIDV')
plot_money_by_day(vcb_df, 'VCB')
plot_money_by_day(vietin_df, 'Vietin')
```

```python
# Tính tổng số tiền cho từng ngân hàng
total_money_by_bank = combined_df.groupby('bank')['money'].sum()

# Vẽ biểu đồ pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_money_by_bank, labels=total_money_by_bank.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Money Transferred by Bank')
plt.show()
```

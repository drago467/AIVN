```python
def add_vectors(vector1, vector2):
  length = len(vector1)
  result = []
  for i in range(length):
    result.append(vector1[i] + vector2[i])
  return result

vector1 = [1, 2, 3]
vector2 = [5, 6, 7]

result = add_vectors(vector1, vector2)
print(result)
```

```python
import os
import cv2
import json

import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
yolo_yaml_path = 'yolo_data/data.yml'
model = YOLO('yolov8s.yaml').load('yolov8s.pt')
```

```python
epochs = 15
imgsz = 640
batch_size = 8
patience = 5
lr = 0.0005

results = model.train(
    data = yolo_yaml_path,
    epochs = epochs,
    imgsz = imgsz,
    batch = batch_size,
    lr0 = lr,
    patience = patience,
    project = 'models',
    name = 'yolov8/detect/train'
)
```

```python
model_path = 'models/yolov8/detect/train/weights/best.pt'
test_img_path = '/content/yolo_data/val/images/vid_4_9620.jpg'

model = YOLO(model_path)
model.predict(test_img_path, save = True)
```
